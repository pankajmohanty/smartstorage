```python
from ValueIteration import ValueIteration, RollOutValueIteration
import torch
checkpoint = torch.load("ffnn_v.pt")

N = checkpoint['N']
M = checkpoint['M']
K = checkpoint['K']
n_episodes=1
eps = 0.0
eps_decay=0.9999

agent = RollOutValueIteration(N, M, K, n_episodes, eps, eps_decay)
agent.ffnn_v.load_state_dict(checkpoint['model_state_dict'])
```

```python
agent.evaluate()
```
