```python
from ValueIteration import ValueIteration, RollOutValueIteration

N = 25
M = (5, 5)
K = 100
n_episodes=500
eps = 0.0
eps_decay=0.9999

agent = RollOutValueIteration(N, M, K, n_episodes, eps, eps_decay)
agent.train()
```

```python
agent.evaluate()
```

```python
import torch
# Saving Model
torch.save({
            'p_vector': agent.env.env.order_p,
            'model_state_dict': agent.ffnn_v.state_dict(),
            'N': N,
            'M': M,
            'K': K,
            }, "ffnn_v.pt")
```
